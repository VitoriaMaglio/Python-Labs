```python
#tuplas são listas imutáveis; denominadas em ()
#listas são mutáveis; denominadas com []
tx = (0,1,1,3,5)

tx.index(3) # mostra a posição que o elemento está na tupla
tx.count(1) # mostra quantas vezes o elemento aparece na tupla

for i in tx:
  print(i)
```

```python
#dicionário = dict; par chave-valor = chave é o acesso e identificação e valor é o dado
pessoa = {
    'nome': 'Joao', 'idade': 25, 'cidade': 'São Paulo'
}
print(pessoa['nome'])
print(pessoa['cidade'])

#JSON (JavaScript Object Notation) é um formato de texto usado para armazenar e trocar dados entre sistemas
#No Python, é fácil converter entre dicionários e JSON. só importar a biblioteca json

pessoa['profissao'] = 'Engenheiro'
pessoa
```

```python
pessoa.keys() # retorna todas as chaves do dicionario
for i in pessoa.keys():
  print(i,pessoa[i])
```

```python
pessoa.get('sexo', 'erro') #busca o que tem no dict e se não estiver contido retorna erro
```
